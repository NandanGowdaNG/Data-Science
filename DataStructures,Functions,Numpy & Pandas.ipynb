```python

my_list = [10, 20, 30, 40, 50]
print(f"Original List: {my_list}")

my_list.append(60)
print(f"After append(60): {my_list}")

my_list.extend([70, 80])
print(f"After extend([70, 80]): {my_list}")

my_list.insert(0, 5)  
print(f"After insert(0, 5): {my_list}")

my_list.remove(40)
print(f"After remove(40): {my_list}")

popped_item = my_list.pop(1)  # Remove item at index 1 (which is 10)
print(f"After pop(1): {my_list} (Popped item: {popped_item})")


index_of_30 = my_list.index(30)
print(f"Index of 30: {index_of_30}")

my_list.append(30) 
count_of_30 = my_list.count(30)
print(f"Count of 30: {count_of_30}")

my_list.sort()
print(f"After sort(): {my_list}")

my_list.reverse()
print(f"After reverse(): {my_list}")

my_list.clear()
print(f"After clear(): {my_list}")
```

```python
num = int(input("Enter a number: "))

if num > 0:
    if num % 2 == 0:
        print(f"{num} is Positive and Even.")
    else:
        print(f"{num} is Positive and Odd.")
elif num == 0:
    print("The number is Zero.")
else:
    print(f"{num} is Negative.")
```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['nandan', 'maddy', 'Charlie', 'Dravid', 'Eshvar', 'Frank', 'Gracy', 'Harsha', 'shivu', 'Jay'],
    'Age': [25, 32, 18, 47, 35, 29, 41, 22, 30, 38],
    'City': ['bengaluru', 'shimoga', 'raichur', 'Hubli', 'Mysuru', 'Navalgunda', 'mandya', 'Chitradurga', 'Hubli', 'Mumbai'],
    'Salary': [60000,  np.nan, 45000, 80000, 75000, 62000, np.nan, 48000, 71000, 85000]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
print("\nDataFrame Info:")
df.info()

print("\nDataFrame Description:")
print(df.describe())
```

```python
name_city_df = df[['Name', 'City']]
print("\nSelected 'Name' and 'City' columns:")
print(name_city_df)
```

```python
df_dropped = df.drop(columns=['City'])
print("\nDataFrame after dropping 'City' column:")
print(df_dropped)
```

```python
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
print(f"\nMean salary used for filling NaNs: {salary_mean}")
print("DataFrame after filling missing 'Salary' values:")
print(df)
```

```python
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```
