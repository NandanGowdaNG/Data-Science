```python
def is_prime(num):
    """Check if a number is prime"""
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):  # check divisors up to sqrt(num)
        if num % i == 0:
            return False
    return True

n = int(input("Enter a number to check if it's prime: "))
print(f"Is {n} a prime number? {is_prime(n)}")
```

```python
# Squares of Even/Odd Numbers
# Choice: Print squares of even numbers between 100 and 200

print("Squares of even numbers from 100 to 200:")
for i in range(100, 201):
    if i % 2 == 0:  # even check
        print(f"{i}^2 = {i**2}")
```

```python
days = 4
room_rate = 2000
consultation_fee = 1500
num_tests = 3
test_rate = 300
medicine_charges = 2400

total_room_charges = days * room_rate
total_lab_charges = num_tests * test_rate
total_before_discount = total_room_charges + consultation_fee + total_lab_charges + medicine_charges

if total_before_discount > 10000:
    discount = 0.10 * total_before_discount
else:
    discount = 0

final_bill = total_before_discount - discount

print(f"Total before discount: ₹{total_before_discount}")
print(f"Discount applied: ₹{discount}")
print(f"Final Bill: ₹{final_bill}")
```
